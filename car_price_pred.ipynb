```python
# Importing Libraries

import numpy as np
import pandas as pd
```

```python
# Loading Dataset
df=pd.read_csv('carprice.csv')
```

```python
df
```

```python
# Checking Null values

df.isnull().sum()
```

```python
# making independent and dependent variables
x=df.drop(['car_ID','CarName','price'],axis=1)
y=df.price
```

```python
df.dtypes
```

```python
# Encoding using ordinal encoder

from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()
```

```python
x[['cylindernumber','enginetype','enginelocation','drivewheel','carbody','fueltype','aspiration','doornumber','fuelsystem']]=oe.fit_transform(x[['cylindernumber','enginetype','enginelocation','drivewheel','carbody','fueltype','aspiration','doornumber','fuelsystem']])
```

```python
x=x.astype(int)
```

```python
# splitting training and testing data

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=32,test_size=0.20)
```

```python
# Importing algorithm

from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train)
```

```python
y_pred=dtr.predict(x_test)
```

```python
# Checking accuracy score

from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred).round(2)
```
